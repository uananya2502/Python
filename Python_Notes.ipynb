# Lexical Analysis in Python

## Definition

Lexical Analysis is the first phase of Python's compilation process.

It scans the source code character by character and converts it into a sequence of **tokens**.

```text
Source Code → Lexical Analysis → Tokens → Parser
```

---

## What is a Token?

A token is the smallest meaningful unit of a program.

### Example

```python
x = 10 + 20
```

### Tokens Generated

```text
NAME(x)
=
NUMBER(10)
+
NUMBER(20)
```

---

## Role of the Lexical Analyzer

The lexical analyzer (also called the lexer or tokenizer):

- Reads source code character by character
- Groups characters into meaningful tokens
- Removes unnecessary whitespace
- Generates special tokens for Python syntax
- Passes tokens to the parser

---

## Types of Tokens

### 1. Keywords

Reserved words that have predefined meanings in Python.

```python
if
else
for
while
class
def
return
```

### 2. Identifiers

Names created by the programmer.

```python
name
student_age
total_marks
```

### 3. Literals

Fixed values used in a program.

```python
10
3.14
"Hello"
True
None
```

### 4. Operators

Symbols used to perform operations.

```python
+
-
*
/
=
==
>=
```

### 5. Delimiters

Symbols used to separate program elements.

```python
()
[]
{}
:
,
.
```

---

## Example of Lexical Analysis

### Source Code

```python
if age >= 18:
    print("Adult")
```

### Tokens Generated

```text
IF
NAME(age)
>=
NUMBER(18)
:
NEWLINE
INDENT
NAME(print)
(
STRING("Adult")
)
NEWLINE
DEDENT
```

---

## Whitespace Handling

### Ignored Whitespace

The following two statements produce the same tokens:

```python
x=10
```

```python
x = 10
```

### Significant Whitespace

Python uses indentation to define code blocks.

```python
if True:
    print("Hello")
```

The lexer generates:

```text
IF
TRUE
:
NEWLINE
INDENT
PRINT
DEDENT
```

---

## Lexical Analysis vs Parsing

| Lexical Analysis | Parsing |
|------------------|----------|
| Works on characters | Works on tokens |
| Produces tokens | Checks syntax |
| First phase | Second phase |
| Done by lexer | Done by parser |

---

## Python Compilation Flow

```text
Source Code
     ↓
Lexical Analysis
     ↓
Tokens
     ↓
Parsing
     ↓
Abstract Syntax Tree (AST)
     ↓
Bytecode
     ↓
Execution
```

---

## Key Points

- Lexical Analysis is the first step in Python compilation.
- It converts source code into tokens.
- Tokens are passed to the parser.
- Most whitespace is ignored.
- Indentation is converted into INDENT and DEDENT tokens.
- The parser uses tokens to verify Python syntax.

---

## One-Line Definition

**Lexical Analysis is the process of converting Python source code into a stream of tokens that can be understood by the parser.**

# 2.3 Names (Identifiers and Keywords)

## Introduction

Names in Python are used to identify variables, functions, classes, modules, and other objects.

A name is also called an **identifier**.

Examples:

```python
x = 10
student_name = "Atul"

def greet():
    pass

class Student:
    pass
```

Here:

- `x`
- `student_name`
- `greet`
- `Student`

---

## Identifier Definition

An identifier is a sequence of characters used to name a variable, function, class, module, or other object.

### Valid Identifiers

```python
name
age
student_name
_marks
value123
```

### Invalid Identifiers

```python
123name
student-name
first name
@class
```

Reasons:

- Cannot start with a digit.
- Cannot contain spaces.
- Cannot contain special symbols (except `_`).
- Cannot be a keyword.

---

## Rules for Naming Identifiers

### 1. Must Begin With

- A letter (A-Z, a-z)
- An underscore (_)
- Unicode letters

```python
name
_age
π
```

- Letters
- Digits
- Underscores

```python
student1
roll_no
data2025
```

```python
name = "Atul"
Name = "Python"

print(name)
print(Name)
```

Output:

```text
Atul
Python
```

`name` and `Name` are different identifiers.

### 4. No Special Characters

Invalid:

```python
student-name
price$
user@email
```

Valid:

```python
student_name
price
user_email
```

Example:

# Reference Counting (CPython)

CPython primarily uses reference counting.

Every object tracks how many references point to it.

```python
a = [1, 2]
b = a
```

Reference count increases.

```python
b = None
```

Reference count decreases.

When the count reaches zero, the object becomes eligible for deletion.

```python
a = []
b = []

a.append(b)
b.append(a)
```

```text
a → b
↑   ↓
└───┘
```

This creates a cycle.

CPython uses a garbage collector to detect and remove such cycles.

# 4.2.2 Resolution of Names

Name resolution is the process Python uses to find the object associated with a name.

Python follows the **LEGB Rule**.

## LEGB Rule

### Local Scope (L)

Variables defined inside the current function.

```python
def func():
    x = 10
    print(x)
```

### Enclosing Scope (E)

Variables in an outer function.

```python
def outer():
    x = 20

    def inner():
        print(x)

    inner()
```

### Global Scope (G)

Variables defined at module level.

```python
x = 30

def func():
    print(x)
```

### Built-in Scope (B)

Built-in names provided by Python.

```python
print(len("abc"))
```

## LEGB Example

```python
x = "Global"

def outer():
    x = "Enclosing"

    def inner():
        x = "Local"
        print(x)

    inner()

outer()
```

Output:

```text
Local
```

Python searches in the order:

1. Local
2. Enclosing
3. Global
4. Built-in

## global Keyword

Used to modify a global variable.

```python
x = 10

def func():
    global x
    x = 20

func()
print(x)
```

Output:

```text
20
```

## nonlocal Keyword

Used to modify a variable in an enclosing scope.

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
```

Output:

```text
20
```

# 4.2.4 Lazy Evaluation

## Definition

Lazy evaluation is a strategy in which expressions are evaluated only when their values are actually needed.

Instead of computing a value immediately, Python delays the computation until it is required.

---

## Why Use Lazy Evaluation?

- Reduces memory usage
- Improves performance
- Avoids unnecessary computations
- Allows handling of large datasets efficiently

---

## Generators and Lazy Evaluation

Generators are the most common example of lazy evaluation in Python.

```python
def numbers():
    for i in range(5):
        yield i
```

The values are generated only when requested.

```python
gen = numbers()

print(next(gen))
print(next(gen))
```

Output:

```text
0
1
```

---

## Generator Expressions

Generator expressions also use lazy evaluation.

```python
squares = (x * x for x in range(5))
```

Values are generated one at a time.

```python
for num in squares:
    print(num)
```

---

## Difference Between List and Generator

### List

```python
nums = [x * x for x in range(1000000)]
```

- Computes all values immediately
- Uses more memory

### Generator

```python
nums = (x * x for x in range(1000000))
```

- Computes values on demand
- Uses less memory

---

## Benefits of Lazy Evaluation

- Efficient memory usage
- Faster startup time
- Useful for large files and datasets
- Supports infinite sequences

# 4.3 Exceptions

## Introduction

An **exception** is an event that occurs during the execution of a program and disrupts the normal flow of instructions.

When an error occurs, Python creates an exception object. If the exception is not handled, the program terminates and displays an error message.

Exceptions provide a structured mechanism for detecting, reporting, and handling runtime errors.

---

## Why Exceptions Are Important

Without exceptions, programs would abruptly terminate whenever an error occurs.

Exceptions allow programs to:

- Detect runtime errors
- Recover from failures
- Separate error-handling code from normal logic
- Improve program reliability and maintainability

---

## Example of an Exception

```python
x = 10
y = 0

print(x / y)
```

Output:

```text
ZeroDivisionError: division by zero
```

Python raises a `ZeroDivisionError` exception because division by zero is undefined.

---

## Exception Lifecycle

The process of exception handling involves three stages:

### 1. Exception Occurs

An error condition is detected.

```python
10 / 0
```

### 2. Exception Is Raised

Python creates an exception object.

```python
ZeroDivisionError
```

### 3. Exception Is Handled

A suitable exception handler processes the exception.

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

---

# The try-except Statement

The `try` block contains code that may generate an exception.

The `except` block handles the exception.

```python
try:
    number = int(input("Enter a number: "))
except ValueError:
    print("Invalid input")
```

---

## Execution Flow

1. Python executes the `try` block.
2. If no exception occurs, all `except` blocks are skipped.
3. If an exception occurs, Python searches for a matching handler.
4. If found, the handler executes.
5. If not found, the exception propagates upward.

---

## Handling Multiple Exceptions

```python
try:
    value = int(input("Enter a number: "))
    result = 10 / value

except ValueError:
    print("Input must be an integer")

except ZeroDivisionError:
    print("Division by zero is not allowed")
```

---

## Handling Multiple Exceptions in One Block

```python
try:
    value = int(input())
except (ValueError, TypeError):
    print("Invalid value")
```

---

## Capturing Exception Objects

```python
try:
    10 / 0

except ZeroDivisionError as e:
    print("Error:", e)
```

Output:

```text
Error: division by zero
```

The variable `e` references the exception object.

---

# The else Clause

The `else` block executes only if no exception occurs.

```python
try:
    x = int(input("Enter a number: "))
except ValueError:
    print("Invalid input")
else:
    print("You entered:", x)
```

---

# The finally Clause

The `finally` block executes regardless of whether an exception occurs.

```python
try:
    file = open("data.txt")
finally:
    file.close()
```

Common uses:

- Closing files
- Releasing resources
- Database cleanup
- Network connection cleanup

---

## Complete Example

```python
try:
    value = int(input("Enter number: "))
    result = 100 / value

except ValueError:
    print("Invalid number")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("Result:", result)

finally:
    print("Execution completed")
```

# 3.3.2 Attribute Access & Descriptors

## Introduction

In Python, every object contains attributes. An attribute may be:

- A data value (such as `name`, `age`, `salary`)
- A method (a function attached to an object)
- A property or computed value

Example:

```python
student.name
student.age
student.display()
```

When Python accesses an attribute, it does not simply read a value from memory. It follows a sophisticated attribute lookup mechanism involving:

- The object's namespace (`__dict__`)
- The class namespace
- The inheritance hierarchy (MRO)
- Descriptor objects
- Special methods such as `__getattribute__()` and `__getattr__()`

This mechanism allows Python to support advanced features like:

- Properties (`@property`)
- Methods
- Class methods
- Static methods
- ORMs such as Django and SQLAlchemy
- Lazy attribute computation
- Attribute validation

---

# Object Namespaces and __dict__

Most Python objects store their attributes in a dictionary called `__dict__`.

Example:

```python
class Student:
    pass


s = Student()

s.name = "Ananya"
s.age = 20

print(s.__dict__)
```

Output:

```text
{
    'name': 'Ananya',
    'age': 20
}
```

The dictionary stores the mapping:

```
attribute name → attribute value
```

---

# Attribute Lookup Process

Consider:

```python
obj.x
```

Python does not immediately look into `obj.__dict__`.

The simplified lookup order is:

```
1. Call __getattribute__()
        |
        V
2. Search data descriptors
        |
        V
3. Search instance __dict__
        |
        V
4. Search non-data descriptors
        |
        V
5. Search class attributes
        |
        V
6. Search parent classes according to MRO
        |
        V
7. Call __getattr__()
        |
        V
8. Raise AttributeError
```

The central method responsible for this process is:

```python
__getattribute__()
```

---

# __getattribute__()

## Definition

`__getattribute__()` is called **every time an attribute is accessed**.

For:

```python
obj.attribute
```

Python internally performs something similar to:

```python
obj.__getattribute__("attribute")
```

---

## Method Signature

```python
object.__getattribute__(self, name)
```

### Parameters

- `self`: The current instance.
- `name`: The name of the attribute being requested.

---

## Example

```python
class Student:

    def __getattribute__(self, name):
        print("Looking for:", name)
        return object.__getattribute__(self, name)


    name = "Ananya"


s = Student()

print(s.name)
```

Output:

```text
Looking for: name
Ananya
```

---

# Infinite Recursion Problem

A common mistake is accessing attributes using `self`.

Wrong:

```python
def __getattribute__(self, name):
    return self.name
```

Why?

Because:

```python
self.name
```

again calls:

```python
__getattribute__()
```

which again tries to access:

```python
self.name
```

This creates an infinite loop:

```
__getattribute__()
       ↓
self.name
       ↓
__getattribute__()
       ↓
self.name
       ↓
...
```

Result:

```text
RecursionError
```

---

# Correct Way

Always delegate to the parent implementation:

```python
object.__getattribute__(self, name)
```

or:

```python
super().__getattribute__(name)
```

---

# __getattr__()

## Definition

`__getattr__()` is called **only when normal attribute lookup fails**.

It acts as a fallback mechanism.

The sequence is:

```
__getattribute__()
          |
          V
Attribute not found
          |
          V
__getattr__()
```

---

## Method Signature

```python
object.__getattr__(self, name)
```

---

## Example

```python
class Student:

    def __getattr__(self, name):
        return f"{name} does not exist"


s = Student()

print(s.age)
```

Output:

```text
age does not exist
```

---

# __getattribute__ vs __getattr__

| Feature | __getattribute__ | __getattr__ |
|---|---|---|
| Called for | Every attribute access | Missing attributes only |
| Position | First step in lookup | Last step before error |
| Complexity | More powerful and dangerous | Simpler and safer |
| Can cause recursion | Yes | Usually no |

# Practical Example: Validation

```python
class Account:

    def __setattr__(self, name, value):

        if name == "balance" and value < 0:
            raise ValueError("Balance cannot be negative")

        object.__setattr__(self, name, value)


acc = Account()

acc.balance = 500
```

# __delattr__()

## Definition

`__delattr__()` is called when an attribute is deleted.

Example:

```python
del obj.name
```

Internally:

```python
obj.__delattr__("name")
```

---

## Method Signature

```python
object.__delattr__(self, name)
```

---

## Example

```python
class Student:

    def __delattr__(self, name):
        print("Deleting", name)

        object.__delattr__(self, name)


s = Student()

s.name = "Ananya"

del s.name
```

Output:

```text
Deleting name
```

# Descriptor Protocol

## Introduction

A descriptor is an object that defines one or more special methods:

```python
__get__()
__set__()
__delete__()
```

Descriptors allow objects to control how other attributes are accessed.

Many Python features are implemented using descriptors:

- `property`
- Methods
- `classmethod`
- `staticmethod`
- ORM fields

---

# Descriptor Methods

## 1. __get__()

### Purpose

Controls what happens when an attribute is read.

Signature:

```python
descriptor.__get__(self, instance, owner)
```

Parameters:

- `self`: Descriptor object.
- `instance`: Instance where the lookup occurred.
- `owner`: The class of the instance.

---

## Example

```python
class Descriptor:

    def __get__(self, instance, owner):
        print("Getting value")
        return 100


class Demo:

    x = Descriptor()


d = Demo()

print(d.x)
```

Output:

```text
Getting value
100
```

---

# 2. __set__()

## Purpose

Controls what happens during assignment.

Example:

```python
obj.x = value
```

calls:

```python
descriptor.__set__(obj, value)
```

---

## Example

```python
class Descriptor:

    def __set__(self, instance, value):
        print("Setting value:", value)


class Demo:

    x = Descriptor()


d = Demo()

d.x = 50
```

Output:

```text
Setting value: 50
```

---

# 3. __delete__()

## Purpose

Controls what happens when an attribute is deleted.

Example:

```python
del obj.x
```

calls:

```python
descriptor.__delete__(obj)
```

---

## Example

```python
class Descriptor:

    def __delete__(self, instance):
        print("Value deleted")


class Demo:

    x = Descriptor()


d = Demo()

del d.x
```

Output:

```text
Value deleted
```

# Data Descriptor vs Non-Data Descriptor

## Data Descriptor

A descriptor implementing:

- `__get__()`
- and either `__set__()` or `__delete__()`

Examples:

- `property`
- Many ORM fields

Data descriptors have higher priority than instance attributes.

---

## Non-Data Descriptor

A descriptor implementing only:

- `__get__()`

Examples:

- Functions (methods)

Instance attributes have higher priority than non-data descriptors.

---

# Descriptor Lookup Priority

For:

```python
obj.x
```

Python searches in this order:

```
1. Data descriptor
2. Instance dictionary (__dict__)
3. Non-data descriptor
4. Class attribute
5. Parent classes (MRO)
6. __getattr__()
7. AttributeError
```

---

# Complete Descriptor Example

```python
class Positive:

    def __get__(self, instance, owner):
        return instance._value


    def __set__(self, instance, value):

        if value <= 0:
            raise ValueError("Value must be positive")

        instance._value = value


    def __delete__(self, instance):
        del instance._value


class Product:

    price = Positive()


p = Product()

p.price = 100

print(p.price)
```

Output:

```text
100
```

# Relationship with @property

A property is actually a descriptor.

Example:

```python
class Person:

    @property
    def age(self):
        return 20
```

Internally, the `property` object implements:

```
__get__()
__set__()
__delete__()
```

# Important Interview and Exam Points

1. `__getattribute__()` runs for every attribute access.
2. `__getattr__()` runs only when an attribute is not found.
3. `__setattr__()` intercepts every attribute assignment.
4. `__delattr__()` intercepts attribute deletion.
5. Descriptors customize attribute access.
6. Data descriptors override instance attributes.
7. Non-data descriptors can be overridden by instance attributes.
8. `property`, methods, `classmethod`, and `staticmethod` are based on descriptors.
9. Incorrect implementations of `__getattribute__()` and `__setattr__()` can cause infinite recursion.
